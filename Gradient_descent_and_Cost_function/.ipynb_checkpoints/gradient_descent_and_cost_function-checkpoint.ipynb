```python
import pandas as pd
import numpy as np
import math
```

```python
def gradient_descent(x,y):
    #define slope and y intercept
    m_current = b_current = 0
    #define iterations
    iterations = 10000
    
    #define learning rate
    learning_rate = 0.0001
    
    #get length of arrays, assumption is that the two arrays are equal in length
    n = len(x)
    
    #set previous cost
    previous_cost = 0
    
    for i in range(iterations):
        #calculate the predicted y values
        y_predicted = m_current * x + b_current
        
        #calculate the cost 
        cost = (1/n)*sum((y-y_predicted)**2)
        
        #calculate m derivative
        dm = -(2/n)*sum(x*(y-y_predicted))
        #calculate b derivative
        db = -(2/n)*sum(y-y_predicted)
        
        #calculate the new slope
        m_current = m_current - learning_rate * dm
        
        #calculate new slope
        b_current = b_current - learning_rate * db
        
        print("M is {}, b is {}, cost is {}, iteration {}".format(m_current, b_current, cost, i))
```

```python
#read the data from csv file
df = pd.read_csv("test_scores.csv")
df
#create arrays
x = np.array(df.math)
y = np.array(df.cs)
print("Values of x is : ", x)
print("Values of y is : ", y)

gradient_descent(x, y)
```
