One hot encoding is used to categorize nominal variables
Nominal variables are variables that can not be ordered.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#read_data from csv
df = pd.read_csv("homeprices.csv")
df
```

```python
#generate dummy variables
dummies = pd.get_dummies(df.town)
dummies
```

```python
#concatenate the dummy variables with the original dataframes
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
#drop some columns
final = merged.drop(['town', 'west windsor'], axis='columns')
final
```

```python
#create a model
model = LinearRegression()
```

```python
#get the x variables (independent variables) by dropping prices column
x = final.drop(['price'], axis='columns')
x
```

```python
#train the model
model.fit(x, final.price)
```

```python
#predict price in robinsville
model.predict([[2800, 0, 1]])
```

```python
#predict price in west windsor
model.predict([[3400, 0, 0]])
```

```python
#calculate how accurate the model is
model.score(x, final.price)
```
