```python
#import neccesary modules
import pandas as pd
```

```python
#read titanic data
df = pd.read_csv("titanic.csv")
df.head()
```

```python
#do data exploration by dropping unnecessary columns
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Embarked','Cabin'], axis='columns', inplace=True)
df.head()
```

```python
#get the y variable(target)
target = df.Survived
inputs = df.drop('Survived', axis='columns')
```

```python
#use dummies for sex column
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
#concatenate the two dataframes
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.head(3)
```

```python
#drop sex column
inputs.drop('Sex', axis='columns', inplace=True)
inputs.head(3)
```

```python
#check if there is any column with NaN values
inputs.columns[inputs.isna().any()]
```

```python
#print first 10 values of age
inputs.Age[:10]
```

```python
#fill missing age values with the mean of age column
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
#use train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
len(x_train)
```

```python
x_train.head(5)
```

```python
len(x_test)
```

```python
#import Naive Bayes
from sklearn.naive_bayes import GaussianNB
#create a model
model = GaussianNB()
#train the model
model.fit(x_train, y_train)
#check the score of the model
model.score(x_test, y_test)
```
