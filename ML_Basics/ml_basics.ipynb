```python
import numpy as np
#get the median

speed = [99,86,87,88,86,103,87,94,78,77,85,86]

x = numpy.median(speed)

print(x)
```

```python
#get the mean
mean_speed = np.mean(speed)
mean_speed
```

```python
#get the mode
from scipy import stats
import math

mode_speed = stats.mode(speed)
print(mode_speed)
```

```python
#get the standard deviation
std = np.std(speed)
std
```

```python
#variance is the square of the variance
speed_var = np.var(speed)
speed_var
```

```python
#find percentiles
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

percentile_75 = numpy.percentile(ages, 75)

percentile_75

#What is the age that 90% of the people are younger than?
ans = numpy.percentile(ages, 90)
ans
```

```python
#create random data

random_data = numpy.random.uniform(0.0, 5.0, 100000)
#use histogram to visualize the data we have generated
import matplotlib.pyplot as plt
plt.hist(random_data, 5)
plt.show()
```

```python
'''Normal data distribution is where values are concentrated around a given value'''
normal_data = numpy.random.normal(5.0, 1.0, 100000)
plt.hist(normal_data, 100)
plt.show()
```

```python
#scatter plot is a diagram where each value in the data is represented by a dot.
x = numpy.random.normal(5.0, 1.0, 1000)
y = numpy.random.normal(10.0, 2.0, 1000)
plt.scatter(x, y)
plt.show()
```

```python
#Linear Regression
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
