```python
import pandas as pd
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
#check methods available
dir(digits)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
    
```

```python
digits.data[:3]
```

```python
#create a 2 dimensional array
df = pd.DataFrame(digits.data)
df
```

```python
#check the target values
digits.target
```

```python
#append the targets to the dataframe
df['target'] = digits.target
df
```

```python
#split the train and test datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), digits.target, test_size=0.2)
```

```python
#ensemble is a term used when using multiple algorithms to predict the outcome
from sklearn.ensemble import RandomForestClassifier
# n_estimators is an optional parameter that specifies the number of trees to use. Default is 10
model = RandomForestClassifier(n_estimators=70) 
#train the model
model.fit(x_train, y_train)
```

```python
#check the score of the model
model.score(x_test, y_test)
```

```python
#get predicted values of y
y_predicted = model.predict(x_test)
```

```python
#confusion matrix helps plot truth on one axis and predicted values on another axis
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
#a better way to view the predicted values vs true values- using seaborn

import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
